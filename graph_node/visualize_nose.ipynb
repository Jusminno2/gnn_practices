```python
import pandas as pd
import matplotlib.pyplot as plt

# データの読み込み
csv_file_path = '../output/nodes_class_data/10037_processed_classified.csv'  # 必要に応じてファイルパスを変更
nodes_df = pd.read_csv(csv_file_path)

# 固定色の割り当て
ROOM_TYPE_COLORS = {
    'living_room': 'skyblue',
    'kitchen': 'orange',
    'bedroom': 'lightgreen',
    'bathroom': 'deeppink',
    'balcony': 'indigo',
    'entrance': 'yellow',
    'dining_room': 'crimson',
    'study_room': 'tan',
    'storage': 'gray',
    'front_door': 'blue',
    'interior_door': 'red',
    'exterior_wall': 'black',
    'interior_wall': 'brown',
    'empty': 'black'
}

# 可視化
plt.figure(figsize=(10, 10))
plt.title("Node Classification Visualization")

# 各クラスごとに色を設定してプロット
for node_class, color in ROOM_TYPE_COLORS.items():
    # クラスごとのデータを取得
    class_nodes = nodes_df[nodes_df['node_class'] == node_class]
    
    if node_class == 'empty':
        # `empty`クラスのみ白抜き
        plt.scatter(class_nodes['x'], class_nodes['y'], edgecolors='lightgrey', facecolors='none', label=node_class, s=30, alpha=0.7)
    else:
        plt.scatter(class_nodes['x'], class_nodes['y'], c=color, label=node_class, s=30, alpha=0.7)

# 凡例を表示
plt.legend(title="Node Class", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.grid(True)
plt.axis("equal")

# グラフの表示
plt.show()
```
